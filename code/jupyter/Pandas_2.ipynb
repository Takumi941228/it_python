```python
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
```

```python
# サンプルは何でもよいですが、例として2019年のJ1リーグ順位表を利用
import webbrowser
website = 'https://ja.wikipedia.org/wiki/2019年のJ1リーグ#順位表'
webbrowser.open(website)
```

```python
# ブラウザ上で表部分をコピーしておくと、クリップボードから読み込み可能
J1_frame = pd.read_clipboard()
```

```python
J1_frame
```

```python
# 列名（カラム）はcolumns
J1_frame.columns
```

```python
J1_frame['チーム']
```

```python
# 列名にスペースがなければ.列名も可能
J1_frame.チーム
```

```python
# 列名を限定して新規にDataFrame作成
DataFrame(J1_frame,columns=['順','チーム','点'])
```

```python
# 存在しない列名を入れた場合
DataFrame(J1_frame,columns=['チーム','勝','敗','得失点'])
```

```python
# 先頭のデータのみを表示
J1_frame.head()
```

```python
#最後のデータのみ表示
J1_frame.tail()
```

```python
# indexを使用し、行の取り出し
J1_frame.iloc[3]
```

```python
# スライスも可能 3から7の手前(6まで)
J1_frame.iloc[3:7]
```

```python
# 明示的にスライスする場合 3から7まで
J1_frame.loc[3:7]
```

```python
# 列全体へ値の代入
J1_frame['テスト']="テスト代入"
```

```python
J1_frame
```

```python
# 行数と一致するとarangeで代入可能
J1_frame["テスト"] = np.arange(10)
J1_frame
```

```python
# SeriesをDataFrameに追加する
test = Series(["test9","test5"],index=[9,5])
```

```python
test
```

```python
J1_frame['test']=test
J1_frame
```

```python
# 列の削除
del J1_frame['test']
J1_frame
```
